```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("/content/flight_price.xlsx")
```

```python
df.head()
```

```python
df["date"]=df["Date_of_Journey"].str.split("/").str[0]
df["month"]=df["Date_of_Journey"].str.split("/").str[1]
df["year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop("Date_of_Journey",inplace=True,axis=1)
```

```python
df.head(2)
```

```python
df["dep_hour"]=df["Dep_Time"].str.split(":").str[0]
df["dep_min"]=df["Dep_Time"].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(lambda x:x.split(" ")[0])
```

```python
df["Arrival_Time"]
```

```python
df["arrival_hour"]=df["Arrival_Time"].str.split(":").str[0]
df["arrival_min"]=df["Arrival_Time"].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df["Additional_Info"].unique()
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop': 0, '2 stops':2, '1 stop':1, '3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head()
```

```python
df["duration_hr"]=df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
df["duration_min"]=df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```

```python
df.head(2)
```

```python
df["Destination"].unique()
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()

ohe.fit_transform(df[["Airline","Source","Destination"]]).toarray()
```

```python
encoded_frame=pd.DataFrame(ohe.fit_transform(df[["Airline","Source","Destination"]]).toarray(),columns=ohe.get_feature_names_out())
```

```python
encoded_frame
```

```python
new_df=pd.concat([df,encoded_frame],axis=1)
```
